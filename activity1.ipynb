<h1> FILLING EMPTY CELLS <h1\>

```python
birds.isnull().sum()
```

```python
birds['bill_length_mm'] = birds['bill_length_mm'].fillna(birds['bill_length_mm'].mean())
```

```python
birds['sex'] = birds['sex'].fillna(birds['sex'].mode()[0])
```
